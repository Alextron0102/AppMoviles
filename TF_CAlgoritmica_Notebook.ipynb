# Estado del arte 

Para completar el objetivo de este trabajo parcial nos basamos en las diferentes técnicas algorítmicas que existen para hacer de este juego una experiencia lo más cercana al original.

Principalmente, el juego se basa en una constante lucha de buscar el camino más cercano al punto de la meta. Actualmente existen una variedades de diferentes algoritmos que sirven para dicho proposito. Algunos de los cuales vamos a mencionar a continuacion:

*   Dijkstra’s Algorithm

Algoritmo de Dijkstra, también llamado algoritmo de los caminos mínimos, escrito por Edsger Dijkstra, un científico de computación. El concepto de este algoritmo se basa en un vértice de origen y uno de destino, unidos por una serie de caminos cuyo peso es siempre positivo, y se busca el camino mínimo entre estos dos vértices, pasando por todos los caminos de por medio y analizando el peso del conjunto de estos caminos, diferenciándolos y tomando siempre el que menor peso tenga.

*   Greedy Best First Search

Este algoritmo se diferencia de otros previamente vistos como el BFS y Dijkstra, en los cuales la frontera se expande en todas las direcciones. Esto puede ser efectivo si se quiere encontrar caminos óptimos a todas las direcciones desde un punto dado. Sin embargo, en el caso de necesitar el camino a un destino único el Manhattan Distance (Heuristic Search) consigue la distancia hasta el destino y de tal manera se continúa explorando primero con el nodo hijo con la distancia menor hacia el destino final, mediante el uso de una cola de prioridad (PriorityQueue()). Debido a que se le da prioridad a los nodos con menor distancia al objetivo la búsqueda del camino se vuelve más eficiente, ya que se ignoran los caminos más lejanos, es decir, que se alejan de nuestro objetivo

*   A* en pathfinding

Este algoritmo encuentra el camino de un punto determinado a otro en una malla o grafo representativo de un ambiente cuadriculado básico.

#Implementación de la funcionaldiad de las barreras dentro del juego:
El algoritmo empleado para que los bots fueran capaces de poner paredes en el tablero funciona de la siguiente manera: Cada bot se va a comparar con el resto en cuanto a cuantos movimientos les son necesarios para llegar a la meta. Si para un bot A el bot B tiene menos pasos para llegar a su destino, bot A intentará poner una pared en el camino de bot B. Para eso primero se fija en que dirección va bot B (ej. derecha a izq, arriba a abajo, etc) para decidir la orientación de la pared. Una vez hecho eso bot A evalúa las posiciones del camino de bot B. Si encuentra un espacio entre dos posiciones camino del bot B, este le pondrá una pared. Bot A no se moverá por ese turno pues puso una pared y se proseguirá a que otro bot haga el mismo análisis. Cabe resaltar, que basta con que un bot tenga menos pasos que dar que el bot que se está comparando con el resto para que le ponga una pared y descarte evaluar a otro bot. Por lo tanto, se tiene cierto comportamiento greedy en el algoritmo ya que no se le está poniendo necesariamente una pared al bot que esté más cerca entre todos de su destino, sino al primero que se encuentre que tenga menos pasos por recorrer.
